```python

import numpy as np

arr = np.array([1,2,3,4])
py_list = [1,2,3,4]

print("arr===>", arr)
print("py_list===>", py_list)
```

Create Arrays of Zeros

```python
np_zeros_arr = np.zeros((3,3))

"""output 3 rows and 3 cols with number zero"""
np_zeros_arr
# print(np_zeros_arr)
```

Create Arrays of Ones

```python
np_ones_arr = np.ones((4,5))

"""output 4 rows and 5 cols with number one"""
np_ones_arr
# print(np_ones_arr)
```

Create Arrays with a Range

```python
np_range_arr = np.arange(1, 40)

"""Output: This will create arr from 1 - 39"""

np_range_arr
# print(np_range_arr)
```
